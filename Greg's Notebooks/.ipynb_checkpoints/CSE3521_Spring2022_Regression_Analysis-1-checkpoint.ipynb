### Step 1.5: What do these attributes actually mean?
| Wine Feature | Description |
| ------------ | ----------- |
| Fixed Acidity | Most acids involved with wine or fixed or nonvolatile (do not evaporate readily) |
| Volatile Acidity | The amount of acetic acid in wine, which at too high of levels can lead to an unpleasant, vinegar taste | 
| Citric Acid | Found in small quantities, citric acid can add ‘freshness’ and flavor to wines |
| Risidual Sugar | The amount of sugar remaining after fermentation stops, it’s rare to find wines with less than 1 gram/liter and wines with greater than 45 grams/liter are considered sweet |
| Chlorides | The amount of salt in the wine |
| Free Sulfur Dioxide | The free form of SO2 exists in equilibrium between molecular SO2 (as a dissolved gas) and bisulfite ion; it prevents microbial growth and the oxidation of wine |
| Total Sulfur Dioxide | total sulfur dioxide: amount of free and bound forms of S02; in low concentrations, SO2 is mostly undetectable in wine, but at free SO2 concentrations over 50 ppm, SO2 becomes evident in the nose and taste of wine |
| Density | the density of water is close to that of water depending on the percent alcohol and sugar content |
| pH | Describes how acidic or basic a wine is on a scale from 0 (very acidic) to 14 (very basic); most wines are between 3-4 on the pH scale (Lower pH = a more tart & sharp wine) | 
| Sulphates | a wine additive which can contribute to sulfur dioxide gas (S02) levels, which acts as an antimicrobial and antioxidant | 
| Alcohol | The percent alcohol content of the wine (obviously) |
| Quality (Output) | A score of the wine's quality |

### Step 3: Multiple linear regression on all the data.

```python
import numpy as np
import statsmodels.api as sm
X = wine_data.drop('quality', axis=1)
X = sm.add_constant(X)
y = wine_data["quality"]

full_model = sm.OLS(y, X)
full_results = full_model.fit()
print(full_results.summary())
```

### Step 3.5: Summary Stats - What does pd.describe tell us about the data?

```python
wine_summary = wine_data.describe()
wine_summary
```

### Step 4: Interpret results so far
According to this data & the results it gives, the features that appear most useful to us (P<=0.05) are the following:
- Volitile Acidity
- Chlorides
- Free Sulfur Dioxide
- Total Sulfur Dioxide
- pH
- Sulphates
- ABV

### That said, what do these actually have to do w/ overall wine quality?
- Well, if volatile acidity of a wine is too high, then the flavor can become bad.
- Chlorides seems to be significant due to the fact that red grades have a high concentration of ions in the skin that seep into the wine during fermentation ([Source](https://www.awri.com.au/wp-content/uploads/2018/08/s1530.pdf#:~:text=The%20higher%20extraction%20of%20chloride,juice%20x%201.7%20%3D%20606))
- The Sulfur Dioxide/Sulphates in the wine is to prevent oxidation & growth of microbes.
- pH is how acidic or basic a wine is, which *could* be related to the volitile acidity content (But I don't know)
- ABV is here for (what I can gather) two reasons. One is that red wines are designed to have high ABVs (at least in comparison to white wine) due to the tannin in the grape skin, and because ABV directly impacts a wine's "body" (Higher ABV = A more "intense" wine) ([Source](https://cellaraiders.com/blogs/news/abv-of-wine))

### Step 4: Perform multiple linear regression on the reduced data we (statistically) care about.

```python
reduced_X = wine_data[['volatile acidity', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'pH', 'sulphates', 'alcohol']]
reduced_X = sm.add_constant(reduced_X)

reduced_model = sm.OLS(y, reduced_X)
reduced_results = reduced_model.fit()
print(reduced_results.summary())
```

### Step 6: Interpret the reduced summary against the full summary
- Going from the full model to this reduced model, we can see the that R-squared decreased by 0.002 and Adj. R-squared increased by 0.001. When we drop data that isn't useful to us, this is the expected result (penalty paying and whatnot).

### Step 7: Use ANOVA to calculate the SSE & F-Statistic.

### Seems like we found a pretty decent model. Now lets make sure its good.

### Step 8: Check that the new model fits the assumptions of linear regression.

### Part 1: Check Assumtion of Linearity

```python
# Borrowed heavily from https://www.kaggle.com/casper06/linear-regression-alcohol-content-prediction
from sklearn.model_selection import train_test_split
useful_wine_data = wine_data[['volatile acidity', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'pH', 'sulphates', 'alcohol', 'quality']]
lin_test_X = useful_wine_data.drop(['quality'], axis=1)
lin_test_y = useful_wine_data['quality']

lin_test_X = sm.add_constant(lin_test_X)
x_train, x_test, y_train, y_test = train_test_split(lin_test_X, lin_test_y, train_size=0.7, random_state=50)

model = sm.OLS(y_train, x_train).fit()
print(model.summary())
```

```python
#plot the actual vs predicted values
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
y_pred = model.predict(x_test)
sns.regplot(y_test, y_pred, line_kws={'color':'red'}, ci=None)

plt.xlabel('Actual')
plt.ylabel('Predictions')
plt.title('Prediction vs Actual')

plt.show()
```

### Part 2: Check Assumption of Multicollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calVIF(X):
    vif = pd.DataFrame()
    vif['variables'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]

    return vif

calVIF(useful_wine_data.loc[:, useful_wine_data.columns != 'quality'])
```

#### From this we can see that the highly-correlated variables in the reduced model is mostly pH and alcohol.

### Part 3: Check Assumption of Homoscedasticity

```python
from statsmodels.stats.diagnostic import normal_ad, het_breuschpagan
#Breusch-Pagan test
bp = het_breuschpagan(model.resid, model.model.exog)
print('Lagrange multiplier statistic: {0} and p-value: {1}'.format(bp[0], bp[1]))
```

#### From these results, we cannot say that the reduced model has any homoscedasticity

### Part 4: Check Assumption of Error Term Normality.

```python
#Anderson-Darling test
def normalityError(residuals):
    p_value = normal_ad(residuals)[1]

    print('p-value from the test:', p_value)

    if p_value < 0.05:
        print('Residuals are not normally distributed')
    else:
        print('Residuals are normally distributed')
        
normalityError(model.resid)
```

### Step 9: Interpreting results from assumption tests.
The reduced model is good, but it fails to meet some of the assumptions of linear regression. The two assumptions that the reduced model didn't fulfill are the normality of error terms, multicollinearity (and maybe linearity?)

Here's what can be done to fix these problems
- For the error terms we need to measure the skewness of the features and transform the columns based on how skewed the distribution is.
- To fix multicollinearity, all we need to do is remove the pH and alcohol features from our model.
Note that these changes should be applied to the FULL model, before we attempt to reduce the model.
